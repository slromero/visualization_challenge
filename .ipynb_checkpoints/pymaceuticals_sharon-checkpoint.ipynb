### Analysis

- The treatments with the lowest final tumor volumes are Capomulin and Ramicane.  Further review would be needed to verify the beginnng volumes against the final tumor volumes to confirm that the treatments reduced the size, not that the tumors were smaller to begin with, for the mice that were treated with Capomulin and Ramicane.
  
- The line chart for the Capomulin treatment of mouse l509 shows an interesting pattern, suggesting that there is some lead time before the treatment begins to take effect. In fact, at the beginning, the tumor appears to grow in size and then drops off around the 20-day mark, increases again around day 25 and then drops in size around day 30, and then again, around day 35, the tumor appears to grow in size.  I would be interested to check whether other mice on Capomulin or the other treatments show a similar pattern, and try to account for it if so.
  
- The correlation coeeficient between mouse weight and average tumor volume is 0.84. Anything over 0.7 typically suggests a strong correlation between two variables, so weight and tumor volume do appear to move together in a somewhat linear fashion: as the mouse weight increases, on average so does the tumor volume. 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = 'data/Mouse_metadata.csv'
study_results_path = 'data/Study_results.csv'

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_study = pd.merge(study_results,mouse_metadata, how='left', on=['Mouse ID', 'Mouse ID'])

# Display the data table for preview
mouse_study.head()
```

```python
# Checking the number of mice.
mouse_metadata['Mouse ID'].count()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
mouse_study_dupes = mouse_study.loc[mouse_study.duplicated(subset=['Mouse ID', 'Timepoint']) == True, 'Mouse ID'].unique()
mouse_study_dupes
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
mouse_study_dupe_df = mouse_study.loc[mouse_study['Mouse ID'] == 'g989',:]
mouse_study_dupe_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
mouse_study_clean = mouse_study.loc[mouse_study['Mouse ID'] != 'g989',:]
mouse_study_clean.head()
```

```python
# Checking the number of mice in the clean DataFrame.
mouse_count = mouse_study_clean['Mouse ID'].value_counts()
len(mouse_count)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
mouse_study_summary = pd.DataFrame()
mouse_study_summary['Mean Tumor Volume'] = mouse_study_clean[['Drug Regimen','Tumor Volume (mm3)']].groupby(['Drug Regimen']).mean()
mouse_study_summary['Median Tumor Volume'] = mouse_study_clean[['Drug Regimen','Tumor Volume (mm3)']].groupby(['Drug Regimen']).median()
mouse_study_summary['Tumor Volume Variance'] = mouse_study_clean[['Drug Regimen','Tumor Volume (mm3)']].groupby(['Drug Regimen']).var()
mouse_study_summary['Tumor Volume Std. Dev.'] = mouse_study_clean[['Drug Regimen','Tumor Volume (mm3)']].groupby(['Drug Regimen']).std()
mouse_study_summary['Tumor Volume Std. Err.'] = mouse_study_clean[['Drug Regimen','Tumor Volume (mm3)']].groupby(['Drug Regimen']).sem()
mouse_study_summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
mouse_study_agg = mouse_study_clean.groupby('Drug Regimen').agg({'Tumor Volume (mm3)' : ['mean', 'median', 'var', 'std', 'sem']})
mouse_study_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
finalTimepointVolume = mouse_study_clean.groupby(['Mouse ID'])['Timepoint'].max()
finalTimepointVolume = finalTimepointVolume.reset_index()
finalTimepointVolume = finalTimepointVolume.rename(columns={'Timepoint':'Final Timepoint'})
finalTimepointVolume
```

```python
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
mouse_study_finalTP = pd.merge(mouse_study_clean,finalTimepointVolume, how='left', on='Mouse ID')
mouse_study_finalTP = mouse_study_finalTP.loc[mouse_study_finalTP['Timepoint'] == mouse_study_finalTP['Final Timepoint']]
mouse_study_finalTP = mouse_study_finalTP.sort_values('Drug Regimen').reset_index()
mouse_study_finalTP
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
finalTumorVol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatments:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    selectedTumorVol = mouse_study_finalTP.loc[mouse_study_finalTP['Drug Regimen']==treatment,'Tumor Volume (mm3)']
    
    # add subset 
    finalTumorVol.append(selectedTumorVol)
    
    # Determine outliers using upper and lower bounds
    quartiles = selectedTumorVol.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)

    outliers = selectedTumorVol.loc[(selectedTumorVol < lower_bound) | (selectedTumorVol > upper_bound)]
    print(f'{treatment} potential outliers: {outliers}')
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
formatted_outlier = dict(markerfacecolor="red", markersize=12)
plt.boxplot(finalTumorVol,labels=treatments,flierprops=formatted_outlier)
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
mouse_l509 = mouse_study_clean.loc[mouse_study_clean['Mouse ID']=='l509',['Timepoint','Tumor Volume (mm3)']]
plt.plot(mouse_l509['Timepoint'],mouse_l509['Tumor Volume (mm3)'])
plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Capomulin treatment of mouse l509')
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
avg_tumor_vol_Cap = mouse_study_clean.loc[mouse_study_clean['Drug Regimen']=='Capomulin',['Mouse ID', 'Tumor Volume (mm3)']].groupby(['Mouse ID']).mean()
avg_tumor_vol_Cap = avg_tumor_vol_Cap.rename(columns={'Tumor Volume (mm3)':'Average Tumor Volume (mm3)'})

mouse_weight = mouse_study_clean[['Mouse ID','Weight (g)']].groupby(['Mouse ID']).mean()

mouse_study_avg_vol = pd.merge(mouse_weight,avg_tumor_vol_Cap, how='right', on = 'Mouse ID').sort_values('Weight (g)')

plt.scatter(mouse_study_avg_vol['Weight (g)'],mouse_study_avg_vol['Average Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Average Tumor Volume by Weight - Capomulin Regimen')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
plt.scatter(mouse_study_avg_vol['Weight (g)'],mouse_study_avg_vol['Average Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Average Tumor Volume by Weight - Capomulin Regimen')
p_slope, p_int, p_r, p_p, p_std_err = st.linregress(mouse_study_avg_vol['Weight (g)'],mouse_study_avg_vol['Average Tumor Volume (mm3)'])
p_fit = p_slope * (mouse_study_avg_vol['Weight (g)']) + p_int
plt.plot(mouse_study_avg_vol['Weight (g)'],p_fit,"red")
print(f"The correlation coefficient between mouse weight and the average tumor volume is {round(st.pearsonr(mouse_study_avg_vol['Weight (g)'],mouse_study_avg_vol['Average Tumor Volume (mm3)'])[0],2)}")
plt.show()
```
